1) Write a python programming to display a bar chart of the popularity of programming languages.
Sample data:
Programing Languages:- Java, Python, PHP, Javascript, C#, C++
Popularity:- 22.2,17.6,8.8,8,7.7,6.7

```python
import matplotlib.pyplot as plt

languages=['Java','Python','PHP','Javascript','C#','C++']
popularity=[22.2,17.6,8.8,8,7.7,6.7]

plt.bar(languages,popularity,width=0.7)
plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.title('Bar Chart')
plt.show()
```

2) Write a python program to create a bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
Sample Data:
Means(men)=(22,30,35,35,26)
Means(women)=(25,32,30,35,29)

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
x1=x-0.3
men=[22,30,35,35,26]
women=[25,32,30,35,29]
labels=['Group1','Group2','Group3','Group4','Group5']
f,a=plt.subplots()
a.bar(x,women,tick_label=labels,label='Women',color='green',width=0.3)
a.bar(x1,men,label='Men',color='blue',width=0.3)
plt.legend()
plt.xlabel('Groups')
plt.ylabel('Mean')
plt.title('Bar Plot of Men and Women')
plt.show()
```

3) Write a python program to add textures(black and white) to bars and wedges.
>Note: Use bottom to stack the women's bars on top of the men's bars.

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
men=[22,30,35,35,26]
women=[25,32,30,35,29]
label=['Group1','Group2','Group3','Group4','Group5']
plt.bar(x,men,tick_label=label,hatch='o',label='Men')
plt.bar(x,women,tick_label=label,hatch='|',bottom=women,label='Women')
plt.legend(loc='upper right') # loc is used to provide the position of the legend which can be given either in form of string or in the form of code correspond to that string. For e.g: here 'upper right' can be replaced by 1.
plt.xlabel('Groups')
plt.ylabel('Mean')
plt.title('Bar Chart of Mean of Men and Women')
plt.show()
```

4) Write a python programming to create a pie chart with a title of the popularity of Programming Languages.
>Sample data:
Programming Languages:- Java, Python, PHP, Javascript, C#, C++
 Popularity:- 22.2,17.6,8.8,8,7.7,6.7

```python
import matplotlib.pyplot as plt

languages=['Java','Python','PHP','Javascript','C#','C++']
popularity=[22.2,17.6,8.8,8,7.7,6.7]
colors=['red','green','blue','yellow','magenta','orange']
plt.pie(popularity,labels=languages,colors=colors,shadow=True,autopct='%1.1f%%',startangle=90)
plt.legend()
plt.title('Popularity of Programming Languages')
plt.show()
```
