```python
# Line Plot
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0.,4,0.1)
y=1+np.cos(2*np.pi*x)
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Cosine Graph')
# grid() is used to show the graph on grid
plt.grid()
plt.show()
```

```python
# Subplot
import matplotlib.pyplot as plt
import numpy as np

x1=np.linspace(0.0,10)
x2=np.linspace(0.0,5)
y1=np.cos(np.pi*x1)
y2=np.cos(np.pi*x2)

plt.subplot(2,1,1) #subplot(211)
plt.plot(x1,y1,'g--')
plt.xlabel('x1')
plt.ylabel('y1')

plt.subplot(2,1,2) #subplot(212)
plt.plot(x2,y2,'o-')
plt.xlabel('x2')
plt.ylabel('y2')

plt.show()
```

```python
# Bar graph is used when the data is analysed on the basis of single axis
# Bar Plot
import matplotlib.pyplot as plt
import numpy as np

x=[1,2,4,5,3]
y=[100,230,536,767,234]
label=['Japan','China','Saudi Arabia','England','France']
colors=['red','blue','green','black','yellow']

plt.bar(x,y,tick_label=label,color=colors,width=0.5) #color='color_name'
plt.xlabel('Countries')
plt.ylabel('Population')
plt.title('Population in Different Countries')
plt.show()
```

```python
# Histogram
import matplotlib.pyplot as plt
temp=[23,47,51,34,20,46,37,49,50,42,34,24,45,35,36,24,38,52]
range=(20,60)
bins=10
plt.hist(temp,bins,range,color='red',rwidth=0.7,histtype='bar')
plt.xlabel('Temperature(in Celsius)')
plt.ylabel('bins')
plt.title('Histogram')
plt.show()
```

```python
# Scatter Plot
import matplotlib.pyplot as plt

x=[1,23,44,13,56,42,46,21,6,5,25,35,74,65]
y=[54,5,23,24,24,67,68,8,32,53,5,74,74,32]
# In scatter() the attribute label is used to give names used by legend() which is basically used when you show different outputs on the same graph
# the attribute s is used to define the size of the marker
plt.scatter(x,y,label='point',color='blue',marker='.',s=50)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot')
# Since we have defined the labels then we should call legend()
plt.legend()
plt.show()
```

```python
# Pie Chart
import matplotlib.pyplot as plt
chocolates=['Dairy Milk','Perk','Munch','5 Star','Galaxy']
quantity=[10,3,5,2,9]
colors=['r','b','g','y','m']
# optional arguments are:-
# startangle used to show from where the piechart starts, shadow takes boolean value used to show 3D effect, explode is used to show the parts seperately, autopct is used to show data in percentage  
plt.pie(quantity,labels=chocolates,colors=colors, startangle=90, shadow=True, explode=(0,0.5,0,0.3,0), autopct='%1.1f%%')
#plt.legend() # It can or cannot be used.
plt.show()
```
