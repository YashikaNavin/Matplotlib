Matplotlib:
Matplotlib is a library in python programming for creating graphs and plots.

Pyplot:
Pyplot is a module in matplotlib which is a collection of functions that are used to create plot easily.

Plotting a simple graph using pyplot:-

```python
# Import the modules:
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Use plot() in pyplot to create a graph or figure which takes different arguments
# Here we have passed an array as argument which it assumes as a sequence of 'Y' values and it automatically generates the 'X' values starting it from 0 by default and having the same length as y.
# So it takes 'X' values as [0,1,2,3]
plt.plot([1,2,3,4])
# Showing what we plotted
plt.show()
```

```python
# Changing the number of arguments in plot(x,y)
plt.plot([1,2,3,4],[15,6,78,8])
```

```python
# Formatting the style of the plot
# Here we have used the 'ro' string which means red dots. By default when no string is passed as argument it takes 'b-' which means blue line. 
plt.plot([1,2,3,4],[15,6,78,8],'ro')
plt.plot([1,2,3,4],[15,6,78,8],'g--')
```

```python
# axis() is a function that takes a list of [xmin, xmax, ymin, ymax] which specifies the view of the axis.
plt.axis([0,6,0,100])
```

```python
# Using the plot() and axis() as well as numpy function to create a graph
a=np.arange(0,10,2)
plt.plot(a,a,'b-',a*2,a,'g^',a*3,a+2,'r--')
plt.axis([0,30,0,100])
# Labelling the graph
plt.title('Normal Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# Controlling the line using 'linewidth' attribute
plt.plot([1,2,3,4],[15,6,78,8],linewidth=5.0)
```

```python
# Controll the size of figure using figsize attribute in figure()
plt.figure(figsize=(3,9))
plt.plot([1,2,3,4],[15,6,78,8],linewidth=5.0)
```
