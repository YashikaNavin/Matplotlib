```python
# Plotting different graphs on a single plot
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0.,2*np.pi,0.1)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label='sin(x)')
plt.plot(x,y2,label='cos(x)',color='m') # color argument is optional
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sine and Cosine Graphs')
plt.legend()
plt.show()
```
